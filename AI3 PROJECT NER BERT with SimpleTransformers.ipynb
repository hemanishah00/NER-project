### Data Prep

```python
wiesp_train_df.head(2)
```

```python
wiesp_train_df.shape
```

```python
wiesp_train_df['unique_id'].nunique()
```

```python
new_data = wiesp_train_df[['ner_ids','ner_tags','section','tokens','unique_id']]
```

```python
new_data = new_data.set_index(['section','unique_id']).apply(pd.Series.explode).reset_index()
```

```python
new_data.isna().sum()
```

```python
new_data["section"] = LabelEncoder().fit_transform(new_data["section"] )
```

```python
new_data["unique_id"] = LabelEncoder().fit_transform(new_data["unique_id"] )
```

```python
new_data.head()
```

```python
new_data.rename(columns={"unique_id":"sentence_id","tokens":"words","ner_tags":"labels"}, inplace =True)
```

```python
new_data["labels"] = new_data["labels"].str.upper()
```

```python
X= new_data[["sentence_id",'section',"words"]]
Y =new_data["labels"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2,random_state=1)
```

```python
#train data and test data
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"section":x_train["section"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"section":x_test["section"],"labels":y_test})
```

```python
train_data.head()
```

### Model

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
label = new_data["labels"].unique().tolist()
#label
```

```python
args = NERArgs()
args.num_train_epochs = 1
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
```

```python
model = NERModel('bert', 'bert-base-cased',labels=label,args =args,)#use_cuda=False)
```

```python
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
result, model_outputs, preds_list = model.eval_model(test_data)
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
example = "My name is Wolfgang and I live in Berlin"

ner_results = nlp(example)
print(ner_results)
```
